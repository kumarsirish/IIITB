Let's import the Pandas and Seaborn libraries, and the `pyplot` module of the Matplotlib library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Let's load and explore the dataset

```python
df = pd.read_csv('churn_data.csv')
df
```

Let's make a pie chart

Here are the counts of entries for different payment methods. This is what we want to depict in our pie chart.

```python
payment_method_value_counts = df['PaymentMethod'].value_counts()
payment_method_value_counts
```

As of version 0.13.2, there is no pie chart method built into Seaborn. Therefore, we will use the method from Matplotlib instead.

```python
plt.figure(figsize = (5, 3))
plt.pie(
    x = payment_method_value_counts.values,
    labels = payment_method_value_counts.index
)
plt.title('Pie chart showing the distribution by payment method')
plt.show()
```

Let's add the percentage for each slice

```python
plt.figure(figsize = (5, 3))
plt.pie(
    payment_method_value_counts.values,
    labels = payment_method_value_counts.index,
    autopct = '%.1f%%'  # percentage for each slice; here %% is an escape sequence for an actual % sign
)
plt.title('Pie chart showing the distribution by payment method')
plt.show()
```

Let's emphasise the 'Bank transfer' slice

```python
plt.figure(figsize = (5, 3))
plt.pie(
    payment_method_value_counts.values,
    labels = payment_method_value_counts.index,
    autopct = '%.1f%%',
    explode = [0, 0, 0.2, 0]  # exploding the bank transfer slice for emphasis
)
plt.title('Pie chart showing the distribution by payment method')
plt.show()
```
