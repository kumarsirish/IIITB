Let's import the Pandas and Seaborn libraries, and the `pyplot` module of the Matplotlib library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's load and explore the data

```python
df = pd.read_csv('churn_data.csv')
df
```

```python
df.info()
```

Let's create a pair plot

A pair plot will show the relations between numeric variables. By default, it will plot scatter plots when comparing two different variables, and a histogram on the top-left to bottom-right diagonal for single variables.

```python
plt.figure(figsize = (5, 3))
sns.pairplot(data = df)  # the function will automatically capture only numeric variables in the dataset
plt.title('Pair plot of numeric variables')
plt.show()
```

```python
# Let's fix the title
plt.figure(figsize = (5, 3))
sns.pairplot(data = df)
plt.suptitle('Pair plot of numerical variables', y = 1.02)
plt.show()
```

Let's customise the plots by changing the marker styles for the scatter plots, and by colouring by the payment method

```python
plt.figure(figsize = (5, 3))
sns.pairplot(data = df, markers = '+', hue = 'Churn')
plt.suptitle('Pair plot of numeric variables', y = 1.02)
plt.show()
```

We can change what plots are displayed using the `kind` and `diag_kind` parameters. For instance, let's make every plot a kernel density estimate (KDE) plot to show a continuous distribution of data

```python
plt.figure(figsize = (5, 3))
sns.pairplot(data = df, kind = 'kde', diag_kind = 'kde')  # changing the plots the KDE
plt.suptitle('Pair plot of numeric variables', y = 1.02)
plt.show()
```
