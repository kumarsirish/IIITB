Let's import the Pandas and Seaborn libraries, and the `pyplot` module of the Matplotlib library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's load and explore the dataset

```python
df = pd.read_csv('churn_data.csv')
df
```

```python
df.info()
```

Let's make a box plot

```python
plt.figure(figsize = (5, 3))
sns.boxplot(data = df, y = 'MonthlyCharges')
plt.title('Boxplot of monthly charges')
plt.ylabel('Monthly charges')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.boxplot(
    data = df,
    x = 'PaymentMethod',
    y = 'MonthlyCharges',
    order = sorted(list(df['PaymentMethod'].unique()))  # sorting boxes alphabetically
)
plt.title('Boxplot of monthly charges for each payment method')
plt.xlabel('Payment method')
plt.xticks(rotation = 20)
plt.ylabel('Monthly charges')
plt.show()
```
