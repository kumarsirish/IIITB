Let's import the NumPy, Pandas, and Seaborn libraries, as well as the `pyplot` module of the Matplotlib library

```python
import numpy as np  # to create a NumPy array
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's load and explore the dataset

```python
df = pd.read_csv('AirPassengers.csv')
df
```

```python
df.info()
```

Let's make a line plot

```python
plt.figure(figsize = (8, 3))
sns.lineplot(data = df, x = 'Month', y = 'Passengers')
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

The ticks on the x-axis are unreadable, because every single month is currently being plotted. Changing this to displaying the dates of only a particular month each year might be beneficial. Let's display only the dates of December in each year

```python
# Let's make the plot again, this time assigning the plot to a variable
ax = sns.lineplot(data = df, x = 'Month', y = 'Passengers')
x_tick_labels = ax.get_xticklabels()
plt.close()  # don't want to display the plot, just want to use the x_tick_labels
```

```python
x_tick_labels
```

We now want to select which of these ticks to display. Let's try to print the labels for December of every two years. Let's start at the 24th tick label (index 23), and proceed yearly until the 144th. We can apply a step of size 24 to display values every two years. Arriving at a solution like this requires experimentation

```python
x_tick_labels_array = np.array(x_tick_labels)  # converting to a NumPy array so we can index it using a range object
x_tick_labels_array[range(23, 144, 24)]
```

We've successfully managed to reduce our labels to just 6. This will make our x-axis tick labels more readable

```python
plt.figure(figsize = (8, 3))
sns.lineplot(data = df, x = 'Month', y = 'Passengers')
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))  # use range() with plt.xticks() to depict only a range of the ticks on the x-axis
plt.ylabel('Passengers')
plt.show()
```

Let's highlight the actual data points in our plot

```python
plt.figure(figsize = (8, 3))
sns.lineplot(data = df, x = 'Month', y = 'Passengers', marker = 'o', markersize = 5, markerfacecolor = 'orange')  # add markers to highlight the individual data points
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))
plt.ylabel('Passengers')
plt.show()
```

We don't want our graph to begin at 100 on the y-axis; we would prefer it begins at 0. We would also like some buffer space at the top. Let's change the y-axis limits to `(0, 700)`

```python
plt.figure(figsize = (8, 3))
sns.lineplot(
    data = df,
    x = 'Month',
    y = 'Passengers',
    marker = 'o',
    markersize = 4,
    markerfacecolor = 'orange',
    linewidth = 1  # reducing line width
)
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))
plt.ylabel('Passengers')
plt.ylim((0, 700))  # setting a range for the y-axis limits
plt.show()
```

In a line plot, the line itself is not real data. It only depicts the connections between real data points. Let's suggest this by changing our line style

```python
plt.figure(figsize = (8, 3))
sns.lineplot(
    data = df,
    x = 'Month',
    y = 'Passengers',
    marker = 'o',
    markersize = 4,
    markerfacecolor = 'orange',
    lw = 1,  # reduce line width
    ls = ':'  # change the style of the line to dotted; can also use parameter name as linestyle
)
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))
plt.ylabel('Passengers')
plt.ylim((0, 700))  # setting a range for the y-axis limits
plt.show()
```
