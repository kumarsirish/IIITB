Let's import the Pandas and Seaborn libraries, and the `pyplot` module of the Matplotlib library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's load and explore the data

```python
df = pd.read_csv('churn_data.csv')
df
```

```python
df.info()
```

Let's calculate the correlations between the numeric variables in our dataset

```python
correlation_matrix = df[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
correlation_matrix
```

We can portraying this information as a heat map

```python
plt.figure(figsize = (7, 5))
sns.heatmap(data = correlation_matrix)
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(data = correlation_matrix, annot = True)  # add annotations to each cell
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(data = correlation_matrix, annot = True, fmt = '.2f')  # formatting the annotations to two decimal places
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(data = correlation_matrix, annot = True, fmt = '.2f', linewidths = 0.5)  # adding space between the cells
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges')
plt.show()
```

Perhaps we would prefer that darker colours be used for higher correlations

```python
plt.figure(figsize = (7, 5))
sns.heatmap(data = correlation_matrix, annot = True, fmt = '.2f', linewidths = 0.5, cmap = 'coolwarm')  # adding a colour map
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.stackplot(data = df, annot = True, fmt = '.2f', linewidths = 0.5, cmap = 'YlGnBu')  # adding a colour map
```
