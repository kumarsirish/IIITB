Data reading functions in Pandas have some useful parameters which we can use to adjust how the data is read into our DataFrame. Let's return to the `pd.read_csv()` function

If, instead of commas, we had another marker separating each data entry in our dataset, we could use the `sep` or `delimiter` parameter to specify this

```python
# As our dataset has no semicolons, all the information of a row will show up in the last column
df1 = pd.read_csv('matches_data.csv', delimiter = ';')
df1.head()
```

By default, Pandas infers our header row from the dataset, typically using the first row. Therefore, by default, `header = 'infer'`. We can change this to another row if we want by specifying a row number

```python
# Here, we're setting the third row as the header, leading to unlikely column names
df2 = pd.read_csv('matches_data.csv', header = 2)
df2.head()
```

We can specify the index if we would like. In the case of our data, the 'id' column could serve as a good index

```python
df3 = pd.read_csv('matches_data.csv', index_col = 'id')
df3.head()
```

```python
df3.index
```

We could also provide multiple columns as an index. This would create a `MultiIndex` object

```python
df4 = pd.read_csv('matches_data.csv', index_col = ['id', 'season'])
df4.head()
```

```python
df4.index
```

We can choose to read only a selection of columns

```python
df5 = pd.read_csv('matches_data.csv', usecols = ['date', 'winner', 'result'])
df5.head()
```

We could also choose to read only a certain number of rows

```python
df6 = pd.read_csv('matches_data.csv', nrows = 3)
df6
```

We could skip a certain number of rows while reading our data

```python
df7 = pd.read_csv('matches_data.csv', skiprows = 500)
df7.head()
```

```python
# We do not want to lose our header row
df8 = pd.read_csv('matches_data.csv', skiprows = range(1, 500))
df8.head()
```
