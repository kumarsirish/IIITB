Let's import the Pandas library

```python
import pandas as pd
```

Let's load our data

```python
df = pd.read_csv('matches_data.csv')
df
```

```python
df.info()
```

We can index and slice DataFrames by position with `.iloc[]`, the integer-location-based (position-based) indexer

```python
df.iloc[0]
```

```python
df.iloc[0, :]
```

```python
type(df.iloc[0, :])  # a single row or column is 1D and therefore a Pandas Series
```

```python
df.iloc[0, 0]
```

```python
df.iloc[:, 0]
```

```python
df.iloc[:2, :5]
```

```python
type(df.iloc[:2, :5])  # a selection of multiple rows and columns is 2D and therefore a Pandas DataFrame
```

```python
df.iloc[::-1, ::-1]  # rows and columns in reverse
```

We can also index and slice DataFrames by using `.loc[]`, the label-based indexer, or just by using regular subscripting

```python
df.loc[0, 'winner']  # here, the label of the row is the same as its numeric index (0)
```

```python
df.loc[:, 'winner': 'target_overs']
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Chandni', 'Dawood'],
    'Age': [25, 30, 18, 22],
    'City': ['Mumbai', 'Kolkata', 'Delhi', 'Bengaluru']
}
df2 = pd.DataFrame(data, index = ["row1", "row2", "row3", "row4"])
df2
```

```python
df2.loc['row1', 'Age']
```

```python
df2.loc[df2.index[0], 'Age']  # you can still use numeric (positional) indexing with the use of the .index[] attribute within .loc[]
```

We can index DataFrames by using conditions

Let's find all the matches where Bengaluru won

```python
df.loc[df['winner'] == 'Bengaluru', :]
```

We want to watch some close, high-scoring games. Let's find the date, teams, target runs, and result margin where the target runs were 200 or more and the winning margin is 10 or less.

```python
df.loc[
    (df['target_runs'] >= 200) & (df['result_margin'] <= 10),
    ['date', 'team1', 'team2', 'target_runs', 'result_margin']
]
```

We're curious about the games where Virat Kohli won player of the match. We also want to know what type of match it was (whether it was a final, semi-final, etc.), and some more information like the date and teams.

```python
df.loc[
    df['player_of_match'].str.contains('Kohli'),
    ['date', 'team1', 'team2', 'player_of_match', 'match_type']
]
```
